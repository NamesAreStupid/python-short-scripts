```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
roidDamage = 0
meleeBoost = 0
damageResist = 0

def normal(level):
    return (20*1.13**level+roidDamage)*(1+meleeBoost)*(1-damageResist)

def krieg(level): 
    return (20*1.13**level*0.8*(1+0.02*level)+roidDamage)*(1+meleeBoost)*(1-damageResist)

x = np.arange(80)
plt.plot(x, normal(x), label='Normal')
plt.plot(x, krieg(x), label='Krieg')
plt.legend(loc='upper left')
plt.show()
```
