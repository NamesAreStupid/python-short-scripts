```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
roidDamage = 0
meleeBoost = 0
damageResist = 0

def normal(level):
    return (20*1.13**level+roidDamage)*(1+meleeBoost)*(1-damageResist)

def krieg(level): 
    level2 = level.copy()
    level2[level2 > 72] = 72
    return (20*1.13**level*0.8*(1+0.02*level2)+roidDamage)*(1+meleeBoost)*(1-damageResist)

x = np.arange(1, 81)
plt.plot(x, normal(x), label='Normal')
plt.plot(x, krieg(x), label='Krieg')
plt.legend(loc='upper left')
plt.show()
```

```python
print('Normal: %.2f' %normal(80))
print('Krieg:  %.2f' %krieg(80))
```

```python
def kriegModifier(level):
    level2 = level.copy()
    level2[level2 > 72] = 72
    return 0.8*(1+0.02*level2)

print(kriegModifier(x))
plt.plot(x, kriegModifier(x))
plt.show()
```
